```python
#Solicite ao usuário um número inteiro positivo.

#Utilizando um laço de repetição, calcule e exiba a soma de
#todos os números de 1 até o número informado.

#Exemplo:

#Digite um número: 5

#Resultado: 15

#Pois:

#1 + 2 + 3 + 4 + 5 = 15

a = int(input("digite um numero "))
for i in range(1,a):
    a += i
print(a)
```

```python
#Solicite ao usuário um número inteiro positivo e calcule
#seu fatorial utilizando um laço de repetição.

#Exemplo:

#Digite um número: 5

#Fatorial = 120

#Pois:

#5! = 5 × 4 × 3 × 2 × 1

#Observação:
#O programa deve validar se o número informado é positivo.

a = int(input("digite um numero "))
try:
  for i in range(1,a):
    a = a * i
  print(a )
except a >= 0:
  print("digite um numero positivo")
```

```python
# Um caixa eletrônico possui notas de:

# R$ 100
# R$ 50
# R$ 20
# R$ 10
# R$ 5
# R$ 2
# R$ 1

# Solicite ao usuário o valor do saque e informe a menor
# quantidade possível de notas para realizar o pagamento.

# Exemplo:

# Digite o valor do saque: 287

# Saída esperada:

# 2 nota(s) de R$100
# 1 nota(s) de R$50
# 1 nota(s) de R$20
# 1 nota(s) de R$10
# 1 nota(s) de R$5
# 1 nota(s) de R$2

# Requisitos:
# - Utilizar laços de repetição.
# - Não utilizar listas.
# - Exibir apenas as notas utilizadas.



def caixa_eletronico():
    try:
        valor = int(input("Digite o valor do saque: "))

        if valor < 0:
            print("Por favor, digite um valor maior que zero.")
            return

        notas = [100, 50, 20, 10, 5, 2, 1]
        print(f"\nSaída esperada:\n")

        for nota in notas:
            quantidade_notas = 0
            while valor >= nota:
                quantidade_notas += 1
                valor -= nota
            if quantidade_notas > 0:
                print(f"{quantidade_notas} nota(s) de R${nota}")

    except ValueError:
        print("Entrada inválida. Por favor, digite um número inteiro.")

# Chamada da função
caixa_eletronico()
```
